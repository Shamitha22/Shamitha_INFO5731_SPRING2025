# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
import math

def hemisphere_volume(ra):
    return (2/3) * math.pi * (ra ** 3)

print("ra\tV")
for ra in range(2, 21, 2):
    V = hemisphere_volume(ra)
    print(f"{ra}\t{V:.2f}")

```

```python

import random

def generate_random_numbers():
    return [random.randint(50, 150) for _ in range(12)]

def compute_average(numbers):
    return sum(numbers) / len(numbers)

# random numbers
random_numbers = generate_random_numbers()
print("Generated Numbers:", random_numbers)

# Sorting in ascending order
ascending_numbers = sorted(random_numbers)
print("Ascending Order:", ascending_numbers)

# Sorting in descending order
descending_numbers = sorted(random_numbers, reverse=True)
print("Descending Order:", descending_numbers)

# Compute and display the average of all numbers
overall_average = compute_average(random_numbers)
print("Overall Average:", round(overall_average, 2))

# Compute and display the average of the first 6 numbers in ascending and descending lists
average_first_6_asc = compute_average(ascending_numbers[:6])
average_first_6_desc = compute_average(descending_numbers[:6])

print("Average of first 6 numbers (Ascending Order):", round(average_first_6_asc, 2))
print("Average of first 6 numbers (Descending Order):", round(average_first_6_desc, 2))


```

['Hello', ',', 'how', 'are', 'you', 'doing', 'today', '?', 'This', 'is', 'an', 'example', 'sentence', 'for', 'the', 'program', '!']

(d) Remove any punctuation from the sentence and print the sanitized sentences. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”).

(e) Count the number of vowels in each of the sentences. Print out each vowel and the occurrences  in the each of the two sentences separately, such as:

Sentence1: a:2, e:2, i:0, etc.
Sentence2: a:1, e:2, i:2, etc.

Design a well-structured Python program to execute these tasks efficiently.

```python

import string

def sentence_to_list(sentence):
    return sentence.split()

def common_words(sentence1, sentence2):
    words1 = set(sentence1.split())
    words2 = set(sentence2.split())
    return list(words1 & words2)

def remove_punctuation(sentence):
    return sentence.translate(str.maketrans('', '', string.punctuation))

def count_vowels(sentence):
    vowels = "aeiouAEIOU"
    vowel_count = {v: sentence.lower().count(v) for v in vowels if v in sentence.lower()}
    return vowel_count

# input
sentence1 = input("Enter the first sentence: ")
sentence2 = input("Enter the second sentence: ")

# sentences into word lists
word_list = sentence_to_list(sentence1) + sentence_to_list(sentence2)
print("Word List:", word_list)

# common words
common = common_words(sentence1, sentence2)
print("Common Words:", common)

# to remove punctuation and print sanitized sentences
sanitized_sentence1 = remove_punctuation(sentence1)
sanitized_sentence2 = remove_punctuation(sentence2)
print("Sanitized Sentence 1:", sanitized_sentence1)
print("Sanitized Sentence 2:", sanitized_sentence2)

# Counting vowels
vowel_count1 = count_vowels(sentence1)
vowel_count2 = count_vowels(sentence2)
print("Vowel Count in Sentence 1:", vowel_count1)
print("Vowel Count in Sentence 2:", vowel_count2)

```

```python
def evaluate_expression(expression):
    try:
        result = eval(expression, {"__builtins__": None}, {"sqrt": math.sqrt, "pow": math.pow})
        return result
    except ZeroDivisionError:
        return "Error: Division by zero is not allowed."
    except Exception as e:
        return f"Error: Invalid expression ({e})"

def main():
    print("Simple Command-Line Calculator")
    print("Supported operations: +, -, *, /, %, ** (exponentiation), sqrt(x)")
    print("Type 'exit' to quit the calculator.")

    while True:
        expression = input("Enter an expression: ")

        if expression.lower() == 'exit':
            print("Exiting calculator. see you later!")
            break

        result = evaluate_expression(expression)
        print("Result:", result)

if __name__ == "__main__":
    main()
```

# Question 6 (20 points).
Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

```python
import csv
import os
from collections import defaultdict

class InventoryManager:
    def __init__(self, filename='inventory.csv'):
        self.filename = filename
        # Edge Case 1: If the CSV file doesn't exist, create one
        if not os.path.exists(filename):
            self.create_inventory_file()

    def create_inventory_file(self):
        """ Creates an empty inventory CSV file with headers if it does not exist """
        headers = ['Product ID', 'Name', 'Category', 'Quantity', 'Price']
        with open(self.filename, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(headers)

    def load_inventory(self):
        """ Loads inventory data from the CSV file, handling corrupted files """
        inventory = []
        try:
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.DictReader(file)
                for row in reader:
                    inventory.append(row)
        except (csv.Error, FileNotFoundError) as e:
            # Edge Case 2: If the file is corrupted or missing, handle gracefully
            print(f"Error loading inventory file: {e}")
        return inventory

    def save_inventory(self, inventory):
        """ Saves inventory data to the CSV file """
        try:
            with open(self.filename, mode='w', newline='') as file:
                writer = csv.DictWriter(file, fieldnames=['Product ID', 'Name', 'Category', 'Quantity', 'Price'])
                writer.writeheader()
                writer.writerows(inventory)
        except csv.Error as e:
            # Edge Case 3: Handle any CSV write errors (e.g., permissions issues)
            print(f"Error saving inventory file: {e}")

    def add_product(self, product_id, name, category, quantity, price):
        """ Adds a new product to the inventory """
        inventory = self.load_inventory()
        # Edge Case 4: Check if product already exists to prevent duplicates
        if any(product['Product ID'] == product_id for product in inventory):
            print(f"Product ID {product_id} already exists.")
            return
        inventory.append({
            'Product ID': product_id,
            'Name': name,
            'Category': category,
            'Quantity': quantity,
            'Price': price
        })
        self.save_inventory(inventory)
        print(f"Product {name} added successfully.")

    def delete_product(self, product_id):
        """ Deletes a product from the inventory """
        inventory = self.load_inventory()
        # Edge Case 5: If the product ID doesn't exist, inform the user
        if not any(product['Product ID'] == product_id for product in inventory):
            print(f"Product ID {product_id} not found.")
            return
        inventory = [product for product in inventory if product['Product ID'] != product_id]
        self.save_inventory(inventory)
        print(f"Product with ID {product_id} deleted successfully.")

    def update_product(self, product_id, name=None, category=None, quantity=None, price=None):
        """ Updates product details """
        inventory = self.load_inventory()
        # Edge Case 6: If the product ID doesn't exist, inform the user
        for product in inventory:
            if product['Product ID'] == product_id:
                if name:
                    product['Name'] = name
                if category:
                    product['Category'] = category
                if quantity is not None:
                    product['Quantity'] = quantity
                if price is not None:
                    product['Price'] = price
                self.save_inventory(inventory)
                print(f"Product {product_id} updated successfully.")
                return
        # Edge Case 6: If product ID is not found, inform user
        print(f"Product ID {product_id} not found.")

    def search_product(self, search_term):
        """ Searches products by name or category """
        inventory = self.load_inventory()
        # Edge Case 7: If no products match the search term, return an empty list
        results = [product for product in inventory if search_term.lower() in product['Name'].lower() or search_term.lower() in product['Category'].lower()]
        if not results:
            print(f"No products found for search term '{search_term}'.")
        return results

    def generate_report(self):
        """ Generates inventory report with item values and category breakdown """
        inventory = self.load_inventory()
        # Edge Case 8: If the inventory is empty, handle gracefully
        if not inventory:
            print("Inventory is empty, no report can be generated.")
            return

        total_value = 0
        category_totals = defaultdict(float)

        for product in inventory:
            try:
                # Edge Case 9: Handle any invalid data (e.g., non-numeric quantity or price)
                value = int(product['Quantity']) * float(product['Price'])
                total_value += value
                category_totals[product['Category']] += value
            except ValueError:
                # Edge Case 9: Log error if quantity or price are invalid
                print(f"Error with product data: {product['Product ID']} - invalid quantity or price.")
                continue

        print(f"Total Inventory Value: ${total_value:.2f}")
        print("Category Breakdown:")
        for category, value in category_totals.items():
            print(f"  {category}: ${value:.2f}")

    def show_inventory(self):
        """ Prints the entire inventory list """
        inventory = self.load_inventory()
        # Edge Case 10: If inventory is empty, inform the user
        if not inventory:
            print("Inventory is empty.")
            return
        for product in inventory:
            print(f"{product['Product ID']}, {product['Name']}, {product['Category']}, {product['Quantity']}, {product['Price']}")

# Example Usage:
if __name__ == "__main__":
    manager = InventoryManager()

    # Add Products (Up to 50 products)
    manager.add_product('001', 'Apple', 'Fruit', 100, 1.20)
    manager.add_product('002', 'Banana', 'Fruit', 150, 0.80)
    manager.add_product('003', 'Laptop', 'Electronics', 30, 999.99)
    # Add more products as needed up to 50...

    # Delete Product
    manager.delete_product('003')

    # Update Product
    manager.update_product('002', quantity=200)

    # Search Product
    results = manager.search_product('Fruit')
    for result in results:
        print(result)

    # Generate Report
    manager.generate_report()

    # Show Inventory
    manager.show_inventory()
```

# Question 7 (20 points)

Create a well desined Python program for a flight booking system. Offer core functions for adding new flights, finding trips, reserving and releasing seating space and creating performance reports. Ensure the system manages edge cases like invalid input, overbooking, and cancellations. Finally, extend the booking system by including advanced features and maintain about 50 flight in the system.

```python
import random
import string

# Flight Class to represent a flight
class Flight:
    def __init__(self, flight_number, destination, available_seats, total_seats=50):
        self.flight_number = flight_number
        self.destination = destination
        self.available_seats = available_seats
        self.total_seats = total_seats
        self.reserved_seats = 0
        self.passenger_list = []

    def reserve_seat(self):
        # #edgecase 1 - Overbooking: If no seats are available, prevent reservation
        if self.available_seats > 0:
            self.available_seats -= 1
            self.reserved_seats += 1
            print(f"Seat reserved successfully for Flight {self.flight_number} to {self.destination}.")
        else:
            print(f"Error: No available seats on Flight {self.flight_number} to {self.destination}.")  #edgecase 1

    def release_seat(self):
        # #edgecase 6 - Canceling a non-reserved seat: Prevent releasing if no seats are reserved
        if self.reserved_seats > 0:
            self.available_seats += 1
            self.reserved_seats -= 1
            print(f"Seat released successfully for Flight {self.flight_number} to {self.destination}.")
        else:
            print(f"Error: No seats to release for Flight {self.flight_number} to {self.destination}.")  #edgecase 6

    def __str__(self):
        return f"Flight {self.flight_number} to {self.destination}: {self.reserved_seats}/{self.total_seats} seats reserved."


# FlightBookingSystem Class to manage the overall booking system
class FlightBookingSystem:
    def __init__(self):
        self.flights = []
        self.booking_records = []

    def add_flight(self, flight_number, destination, available_seats=50):
        # #edgecase 4 - Invalid input for flight details (invalid available seats or flight number)
        if available_seats <= 0 or not flight_number:  # Ensure seats are positive and flight number exists
            print(f"Error: Invalid input for Flight {flight_number}. Available seats must be greater than 0.")
            return

        flight = Flight(flight_number, destination, available_seats)
        self.flights.append(flight)
        print(f"Flight {flight_number} to {destination} has been added with {available_seats} seats.")  #edgecase 4

    def find_flights(self, destination):
        # #edgecase 8 - Searching for flights to a non-existent destination
        available_flights = [flight for flight in self.flights if flight.destination.lower() == destination.lower() and flight.available_seats > 0]

        if not available_flights:
            print(f"No available flights to {destination}.")  #edgecase 8
        else:
            for flight in available_flights:
                print(flight)

    def reserve_seat(self, flight_number):
        # #edgecase 2 - Invalid flight number (Flight does not exist)
        flight = self.get_flight(flight_number)
        if flight:
            flight.reserve_seat()
            self.booking_records.append(flight_number)

    def release_seat(self, flight_number):
        # #edgecase 7 - Releasing a non-reserved seat
        flight = self.get_flight(flight_number)
        if flight:
            flight.release_seat()
            # Only remove the flight number from booking records if it's there (prevents ValueError)
            if flight_number in self.booking_records:
                self.booking_records.remove(flight_number)

    def get_flight(self, flight_number):
        # #edgecase 2 - Invalid flight number (Flight does not exist)
        for flight in self.flights:
            if flight.flight_number == flight_number:
                return flight
        print(f"Error: Flight {flight_number} not found.")  #edgecase 2
        return None

    def generate_report(self):
        print("\n--- Flight Booking System Performance Report ---")
        total_bookings = len(self.booking_records)
        total_cancellations = len([flight for flight in self.flights if flight.reserved_seats == 0])
        available_seats = sum([flight.available_seats for flight in self.flights])

        print(f"Total Bookings: {total_bookings}")
        print(f"Total Cancellations: {total_cancellations}")
        print(f"Total Available Seats: {available_seats}")

        if not self.flights:
            print("No flights available in the system.")  # #edgecase 10 - Empty flight data scenario
        else:
            for flight in self.flights:
                print(flight)

    def handle_edge_cases(self):
        # #edgecase 1 - Overbooking: Check if any flight has more reserved seats than total seats
        for flight in self.flights:
            if flight.reserved_seats > flight.total_seats:
                print(f"Error: Overbooking detected for Flight {flight.flight_number}.")  #edgecase 1

    def extend_features(self):
        # #edgecase 9 - Invalid flight status update
        statuses = ["On Time", "Delayed"]
        for flight in self.flights:
            flight.status = random.choice(statuses)
            print(f"Flight {flight.flight_number} to {flight.destination} is currently {flight.status}.")

    def handle_invalid_input(self):
        # #edgecase 5 - Full flight reservation
        for flight in self.flights:
            if flight.available_seats == 0:
                print(f"Error: Flight {flight.flight_number} to {flight.destination} is already full.")  #edgecase 5


# Main program
if __name__ == "__main__":
    # Initialize the booking system
    booking_system = FlightBookingSystem()

    # #edgecase 4 - Invalid input for flight details
    booking_system.add_flight("AI202", "New York", 50)
    booking_system.add_flight("AI303", "London", 30)
    booking_system.add_flight("AI404", "Paris", 40)

    # #edgecase 8 - Searching for flights to a non-existent destination
    booking_system.find_flights("Tokyo")

    # Reserve seats
    booking_system.reserve_seat("AI202")
    booking_system.reserve_seat("AI303")
    booking_system.reserve_seat("AI404")

    # #edgecase 6 - Canceling a non-reserved seat
    booking_system.release_seat("AI303")
    booking_system.release_seat("AI303")  # Attempting to release a seat that wasn't reserved

    # #edgecase 5 - Full flight reservation (Trying to reserve after all seats are taken)
    booking_system.reserve_seat("AI202")
    booking_system.reserve_seat("AI202")
    booking_system.reserve_seat("AI202")
    booking_system.reserve_seat("AI202")  # AI202 now full

    # #edgecase 4 - Invalid input for flight details
    booking_system.add_flight("AI505", "Berlin", -5)  # Invalid number of available seats

    # #edgecase 10 - Empty flight data on system startup (System initialized with no data)
    empty_system = FlightBookingSystem()
    empty_system.find_flights("London")

    # Generate Performance Report
    booking_system.generate_report()

    # Handle edge cases like overbooking
    booking_system.handle_edge_cases()

    # Extend features: Flight Status
    booking_system.extend_features()
```
